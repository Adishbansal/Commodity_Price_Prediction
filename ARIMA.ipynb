```python
missing_df =  df.isnull().sum().to_frame().rename(columns={0:"Total No. of Missing Values"})
missing_df["% of Missing Values"] = round((missing_df["Total No. of Missing Values"]/len( df))*100,2)
missing_df
```

```python
# Handling missing values
# Imputing missing values with the mean for continuous variables and mode for categorical variables
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].mean(), inplace=True)

# Checking for missing values before imputation
missing_values = df.isnull().sum()

# Rechecking for missing values after imputation
missing_values_after = df.isnull().sum()

(missing_values, missing_values_after)
```

```python
df['Reported Date'] = pd.to_datetime(df['Reported Date'], errors='coerce')
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df_grouped = df.groupby('Reported Date')['Modal Price (Rs./Quintal)'].mean().reset_index()
df_grouped.rename(columns={'Reported Date': 'Date', 'Modal Price (Rs./Quintal)': 'Modal_Price'}, inplace=True)
df_grouped.set_index('Date', inplace=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df_grouped.index, df_grouped['Modal_Price'], marker='o', linestyle='-', color='b')
plt.xlabel('Date')
plt.ylabel('Modal Price (Rs./Quintal)')
plt.title('Modal Price over Time')
plt.grid()
plt.show()
```

```python
rolling_mean = df_grouped['Modal_Price'].rolling(window=12).mean()
rolling_std = df_grouped['Modal_Price'].rolling(window=12).std()

plt.figure(figsize=(12, 6))
plt.plot(df_grouped['Modal_Price'], color='blue', label='Original')
plt.plot(rolling_mean, color='red', label='Rolling Mean')
plt.plot(rolling_std, color='black', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean and Standard Deviation')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(df_grouped['Modal_Price'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
for key, value in result[4].items():
    print('Critical Value {}: {:.3f}'.format(key, value))
```

```python
df_diff = df_grouped['Modal_Price'].diff().dropna()

plt.figure(figsize=(12, 6))
plt.plot(df_diff, color='blue')
plt.title('First Difference of Modal Price')
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Assuming (p,d,q) = (1,1,1) for ARIMA
model = ARIMA(df_grouped['Modal_Price'], order=(1, 1, 1))
model_fit = model.fit()
print(model_fit.summary())
```

```python
forecast = model_fit.forecast(steps=10000)
print(forecast)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_grouped.index, df_grouped['Modal_Price'], label='Historical Data', color='blue')
plt.plot(forecast.index, forecast, label='Forecasted Data', color='red')
plt.xlabel('Date')
plt.ylabel('Modal Price (Rs./Quintal)')
plt.title('Modal Price Forecast')
plt.legend()
plt.show()
```
